```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

## 1. Load Data

```python
df = pd.read_csv('../dataset_raw.csv')
df.head()
```

## 2. Exploratory Data Analysis (EDA)

## 3. Preprocessing

```python
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df[['age', 'income']])
df_preprocessed = pd.DataFrame(scaled_features, columns=['age_scaled', 'income_scaled'])
df_preprocessed['target'] = df['target']

df_preprocessed.to_csv('dataset_preprocessing.csv', index=False)
print("Preprocessing selesai.")
```
